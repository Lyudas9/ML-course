```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import pandas as pd

medical_df = pd.read_csv('medical-charges.csv')
non_smoker_df = medical_df[medical_df.smoker == 'no']
```

```python
non_smoker_df
```

```python
X = non_smoker_df['age'].values.reshape(-1, 1)
X_frame = non_smoker_df['age'].to_frame()
y = non_smoker_df['charges'].values
```

```python
X
```

**Метод МНК** для прогнозу charges за age для некурців

```python
def normal_equations(X,y):
    """Computes the closed-form solution to linear regression"""
    theta = np.linalg.inv(X.T.dot(X)).dot(X.T).dot(y)
    return theta
```

```python
# Додаємо колонку одиничок аби задачи bias терм (константу в моделі регресії)
X_with_intercept = np.c_[np.ones((X.shape[0], 1)), X]
X[:3], X_with_intercept[:3]
```

```python
MNK_coefs = normal_equations(X_with_intercept, y)
```

```python
print("Коефіцієнти моделі МНК:", MNK_coefs[0], ",", MNK_coefs[1])
```

```python
predictions_MNK = np.dot(X_with_intercept, MNK_coefs)
```

```python
predictions_MNK
```

```python
# Візуалізація результатів
plt.figure(figsize=(10, 6))
plt.scatter(X[:, 0], y, color='blue', label='Data points')
plt.plot(X[:, 0], predictions_MNK, color='red', label='MNK')
plt.xlabel('Значення змінної')
plt.ylabel('Цільове значення')
plt.legend()
plt.grid(True)
plt.show()
```

```python
rmse = mean_squared_error(y, predictions_MNK, squared=False)
print('RMSE МНК: ', rmse)
```

**Full-Batch градієнтного спуску** для прогнозу charges за age для некурців

```python
# Градієнтний спуск
def full_batch_gradient_descent(X, y, lr=0.0001, epochs=100):
    m, b = 0.0, 0.0  # Початкові параметри
    n = len(y)
    errors = []
    for _ in range(epochs):
        y_pred = m * X[:, 0] + b
        error = y_pred - y
        errors.append(sum(error**2))
        m_gradient = (2/n) * np.dot(error, X[:, 0])
        b_gradient = (2/n) * np.sum(error)
        m = m - lr * m_gradient
        b = b - lr * b_gradient
    return m, b, errors
```

```python
m, b, errors = full_batch_gradient_descent(X, y, lr=0.01, epochs=1000)
print("Коефіцієнти Full-Batch градієнтного спуску при lr=0.01, epochs=1000: ", m, b)
```

```python
m, b, errors = full_batch_gradient_descent(X, y, lr=0.001, epochs=100)
print("Коефіцієнти Full-Batch градієнтного спуску при lr=0.001, epochs=100: ", m, b)
```

```python
m, b, errors = full_batch_gradient_descent(X, y, lr=0.001, epochs=1000)
print("Коефіцієнти Full-Batch градієнтного спуску при lr=0.001, epochs=1000: ", m, b)
```

**Спостереження:** бачимо, що великий learning rate не дає зійтись нашому градієнту і знаходить якийсь, тому використовуємо lr=0.0001, а також велику кількість епох, щоб градієнтний спуску втиг навчитись достатньо, щоб зійтись.

```python
m, b, errors = full_batch_gradient_descent(X, y, lr=0.0001, epochs=1000)
predictions_fbgd = m * X[:, 0] + b
```

```python
print("Коефіцієнти Full-Batch градієнтного спуску при lr=0.0001, epochs=1000: ", m, b)
```

```python
predictions_fbgd
```

```python
plt.plot(errors);
```

```python
# Візуалізація результатів
plt.figure(figsize=(12, 6))
plt.scatter(X[:, 0], y, color='blue', label='Data points')
plt.plot(X[:, 0], predictions_fbgd, color='green', label='Full-batch Gradient Descent')
plt.title('Gradient Descent')
plt.xlabel('Значення ознаки')
plt.ylabel('Цільове значення')
plt.legend()
plt.grid(True)
plt.show()
```

```python
rmse_fbgd = mean_squared_error(y, predictions_fbgd, squared=False)
print('RMSE градієнтного спуску: ', rmse_fbgd)
```

Виведіть
- точність моделі
-  коефіцієнти
-  візуалізуйте модель у вигляді лінії на графіку розсіювання `age` проти `charges`

і зробіть висновки, чи це хороша модель, чи ви б її використовували в компанії?

```python
smoker_df = medical_df[medical_df.smoker == 'yes']

X_sm = smoker_df['age'].to_frame()
y_sm = smoker_df['charges'].values
```

```python
lin_reg = LinearRegression()
lin_reg.fit(X_sm, y_sm)
print("Коефіцієнти моделі лінійної регресії:", lin_reg.coef_, ",", lin_reg.intercept_)
```

```python
predictions_sklearn_sm = lin_reg.predict(X_sm)
```

```python
X_sm = np.expand_dims(smoker_df['age'].to_numpy(), axis=1)

# Візуалізація результатів
plt.figure(figsize=(12, 6))
plt.scatter(X_sm[:, 0], y_sm, color='blue', label='Data points')
plt.plot(X_sm[:, 0], predictions_sklearn_sm, color='grey', label='Sklearn')
plt.title('Linear Regression to predict charges based on age')
plt.xlabel('Значення ознаки')
plt.ylabel('Цільове значення')
plt.legend()
plt.grid(True)
plt.show()
```

```python
r2 = r2_score(y_sm, predictions_sklearn_sm)
rmse_sklearn_sm = mean_squared_error(y_sm, predictions_sklearn_sm, squared=False)
mae = mean_absolute_error(y_sm, predictions_sklearn_sm)

print(f"R² score: {r2:.2f}")
print(f"Root Mean Squared Error: {rmse_sklearn_sm:.2f}")
print(f"Mean Absolute Error: {mae:.2f}")
```

```python
np.mean(y_sm), np.std(y_sm), np.min(y_sm)
```

**Висновки**:
По значеннях метрик видно, що модель показує погані результати і не може спрогнозувати кількість медичних зборів для курців на основі віку. R² менший за 80%, rmse метрика близька до стандартого відхилення по сумах медичних зборів. Тому таку модель точно не можна використовувати.
