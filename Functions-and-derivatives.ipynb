1. Задана функція залежності оцінки від кількості годин, витрачених на навчання: $ f(x) = 10x - 5 $.
  1. З допомогою Python обчисліть похідну цієї функції.
  2. Використовуйте бібліотеку `matplotlib` для побудови графіків функції та її похідної в діапазоні значень від 0 до 10 годин. Проаналізуйте, як зміна кількості годин вивчення впливає на оцінку.



```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp
```

```python
# Визначаємо символічну змінну
x = sp.symbols('x')

# Лінійна функція
f_lin = 10 * x - 5
derivative_f_lin = sp.diff(f_lin, x)

print("Похідна лінійної функції 10x - 5:", derivative_f_lin)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Визначення функції та її похідної
x = np.linspace(0, 10, 100)
f = 10 * x - 5
df = 10 * np.ones_like(x)  # Похідна як константа у вигляді масиву

# Створення графіку
plt.figure(figsize=(10, 5))
plt.plot(x, f, label='f(x) = 10*x - 5', color='blue')
plt.plot(x, df, label="f'(x) = 10", color='green')

# Позначення точки та дотичної лінії
x_point = 3
f_point = 10 * x_point - 5  # Виправлення розрахунку f_point
df_point = 10
tangent_line = df_point * (x - x_point) + f_point  # Рівняння дотичної: y - y1 = m(x - x1)

plt.scatter(x_point, f_point, color='red')  # точка на кривій
plt.plot(x, tangent_line, '--', color='red', label='Tangent at x = 3')  # дотична

# Додавання легенди, міток та заголовка
plt.title("Графік функції 10*x - 5 та її похідної")
plt.xlabel("Кількість годин (x)")
plt.ylabel("Оцінка / Похідна (y)")
plt.legend()
plt.grid(True)

plt.show()
```

2. Використовуючи Python і `SymPy`, обчисліть похідні для функцій $ f(x) = \sin(x) $ та $ f(x) = \ln(x) $. Зобразіть ці функції та їхні похідні за допомогою `matplotlib`.



```python
# Визначаємо символічну змінну
x = sp.symbols('x')
```

```python
# Рахуємо похідну для sin(𝑥)
f_sin = sp.sin(x)
derivative_f_sin = sp.diff(f_sin, x)

print("Похідна лінійної функції sin(𝑥):", derivative_f_sin)
```

```python
# Визначення змінної
x_sym = sp.symbols('x')

# Визначення функції та її похідної
f_sym = sp.sin(x_sym)
df_sym = sp.diff(f_sym, x_sym)

# Перетворення функцій у NumPy функції
f_np = sp.lambdify(x_sym, f_sym, 'numpy')
df_np = sp.lambdify(x_sym, df_sym, 'numpy')

# Діапазон значень для графіків
x = np.linspace(0, 10, 400)

# Обчислення значень функцій та їх похідних
f = f_np(x)
df = df_np(x)

# Створення графіку
plt.figure(figsize=(10, 5))
plt.plot(x, f, label='f(x) = sin(x)', color='blue')
plt.plot(x, df, label="f'(x) = cos(x)", color='green')


# Додавання легенди, міток та заголовка
plt.title("Графік функції sin(x) та її похідної")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)

plt.show()
```

```python
# Визначаємо символічну змінну
x = sp.symbols('x')

# Рахуємо похідну для ln(𝑥)
f_ln = sp.ln(x)
derivative_f_ln = sp.diff(f_ln, x)

print("Похідна лінійної функції ln(𝑥):", derivative_f_ln)
```

```python
# Визначення змінної
x_sym = sp.symbols('x')

# Визначення функції та її похідної
f_sym = sp.ln(x_sym)
df_sym = sp.diff(f_sym, x_sym)

# Перетворення функцій у NumPy функції
f_np = sp.lambdify(x_sym, f_sym, 'numpy')
df_np = sp.lambdify(x_sym, df_sym, 'numpy')

# Діапазон значень для графіків
x = np.linspace(0.1, 10, 400)

# Обчислення значень функцій та їх похідних
f = f_np(x)
df = df_np(x)

# Створення графіку
plt.figure(figsize=(10, 5))
plt.plot(x, f, label='f(x) = ln(x)', color='blue')
plt.plot(x, df, label="f'(x) = 1/x", color='green')


# Додавання легенди, міток та заголовка
plt.title("Графік функції ln(x) та її похідної")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)

plt.show()
```


3. Загальні витрати компанії від обсягу виробництва задані функцією $ f(x) = x^2 - 5x + 1 $, де $ x $ — кількість вироблених одиниць. Обчисліть похідну цієї функції з допомогою SymPy та візуалізуйте значення функції і її похідної на відрізку від 0 до 100.

```python
# Квадратична функція
f_sq = x_sym**2 - 5*x_sym + 1
derivative_f = sp.diff(f_sq, x_sym)

print("Похідна квадратичної функції:", derivative_f)
```

```python
# Визначення змінної
x_sym = sp.symbols('x')

# Визначення функції загальних витрат
f_sym = x_sym**2 - 5*x_sym + 1

# Обчислення похідної
df_sym = sp.diff(f_sym, x_sym)

# Перетворення функцій у NumPy функції
f_np = sp.lambdify(x_sym, f_sym, 'numpy')
df_np = sp.lambdify(x_sym, df_sym, 'numpy')

# Діапазон значень для графіків
x = np.linspace(0, 100, 400)

# Обчислення значень функцій та їх похідних
f = f_np(x)
df = df_np(x)

# Створення графіку
plt.figure(figsize=(12, 6))
plt.plot(x, f, label='f(x) = x^2 - 5x + 1', color='blue')
plt.plot(x, df, label="f'(x) = 2x - 5", color='green')

# Додавання легенди, міток та заголовка
plt.title("Графік функції загальних витрат та її похідної")
plt.xlabel("Кількість вироблених одиниць (x)")
plt.ylabel("Витрати / Похідна (y)")
plt.legend()
plt.grid(True)

plt.show()
```
